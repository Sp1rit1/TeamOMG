```python
from facenet_pytorch import InceptionResnetV1
from PIL import Image
import numpy as np
import torch
import cv2
```

```python
model = InceptionResnetV1(pretrained='vggface2').eval()
```

```python
def preprocess_face(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Не удалось загрузить изображение: {image_path}")
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (160, 160))
    img = Image.fromarray(img)

    img_tensor = torch.tensor(np.array(img) / 255.0, dtype=torch.float32)
    img_tensor = img_tensor.permute(2, 0, 1).unsqueeze(0)  # [1, 3, 160, 160]
    
    return img_tensor
```

```python
def get_embedding(img_tensor):
    with torch.no_grad():
        embedding = model(img_tensor)
    return embedding[0].numpy()
```

```python

image_path = "img1.png"


img_tensor = preprocess_face(image_path)
embedding = get_embedding(img_tensor)


print("Эмбеддинг:", embedding)
print("Размер эмбеддинга:", embedding.shape)
```
